```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('spotify.csv')
df.set_index('Unnamed: 0', inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

## Limpeza dos dados

```python
# Arrumando dados com unidades esquisitas
df['acousticness'] = df['acousticness'].str.rstrip('kg')
df['danceability'] = df['danceability'].str.rstrip('mol/L')
```

```python
df.head()
```

Notei que ainda existem valores faltantes

```python
df[df == 'nao_sei'].count()
```

```python
# Transformando objetos em n√∫meros
cols = ['song_popularity', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'audio_mode', 'speechiness', 'tempo', 'time_signature']
df[cols] = df[cols].apply(pd.to_numeric, errors = 'coerce')
df['song_duration_ms'] = df['song_duration_ms'].apply(pd.to_timedelta, errors = 'coerce')
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

## Plotando os dados

```python
# Distribui√ß√£o da popularidade das m√∫sicas
plt.figure(figsize=(10,9))
sns.distplot(df['song_popularity'])
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.heatmap(df.corr(), annot=True, vmin=-1, vmax=1)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.pairplot(df[['song_name', 'song_popularity', 'song_duration_ms', 'acousticness',
       'danceability', 'energy', 'instrumentalness']].sample(1000))
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.pairplot(df[['key', 'liveness',
       'loudness', 'audio_mode', 'speechiness', 'tempo', 'time_signature',
       'audio_valence']].sample(1000))
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.scatterplot(data = df, x = 'energy', y = 'loudness')
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.lmplot(data = df, x = 'energy', y = 'loudness', scatter_kws={'alpha':0.005}, order=3, col = 'audio_mode')
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.lmplot(data = df, x = 'acousticness', y = 'energy', scatter_kws={'alpha':0.01}, order=2, col = 'audio_mode')
plt.show()
```
